```python
#install panda sql to simulate dbms
# !pip install pandasql
import pandas as pd
from pandasql import sqldf
pd.set_option('display.max_rows', None)
```

1. Read recently created csv files extracted from json & perfrom data exploration


```python
df = pd.read_csv("taxi.csv", sep=",")
df.head()
df.info()
# Rename the columns for easy querying
df.rename(columns={"geometry.type":"geometry_type",
                   "geometry.coordinates":"coordinates",
                   "properties.timestamp":"timestamp",
                   "properties.taxi_count":"taxi_count",
                   "properties.api_info.status":"status"},
          inplace=True)
df.info()
```

2. Query data from dataframe

```python
# setup our query to view our data
select_query = 'SELECT * FROM df LIMIT 10'
df_select = sqldf(select_query)
display(df_select)
```

3. Select top 10 highest population

```python
select_top_ten = 'SELECT * FROM df ORDER BY taxi_count DESC LIMIT 10'
df_top_ten = sqldf(select_top_ten)
display(df_top_ten)
```

4. Select Top 10 areas of lowest taxi population

```python
select_top_ten_low = 'SELECT * FROM df ORDER BY taxi_count asc LIMIT 10'
df_ten_lowest = sqldf(select_top_ten_low)
display(df_ten_lowest)
```

5. Hourly Taxi avialability

```python
# time format is by hourly therefore %h is used
time_format = "'%h'"
select_hour = f"SELECT * FROM df GROUP BY strftime ({time_format},timestamp)"
#print(select_hour)
df_hour = sqldf(select_hour)
display(df_hour)
```
